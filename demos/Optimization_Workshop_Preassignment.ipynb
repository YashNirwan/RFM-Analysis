```python
import gurobipy as gp
from gurobipy import Env, Model, GRB

env = Env(empty=True)
env.setParam("WLSACCESSID", "94ce3347-9c14-445d-bffe-d9f1e08c7ca1")  # Replace with your Access ID
env.setParam("WLSSECRET", "300e38b7-abbf-43a2-ad68-e94b79d27efb")  # Replace with your Secret Key
env.setParam("LICENSEID", 2810074)  # Replace the number with your license ID
env.start()  # Start the environment
```

If your installation is correct, you should obtain a numerical objective value afer running the cell below!

```python
# Now, create the optimization model using the custom environment
m = Model("example", env=env)

# Define variables
x = m.addVar(name="x")
y = m.addVar(name="y")

# Set objective: Maximize x + y
m.setObjective(x + y, GRB.MAXIMIZE)

# Add constraints
m.addConstr(x + 2*y <= 4, "c1")
m.addConstr(2*x + y <= 3, "c2")

# Solve the model
m.optimize()

# Print the results
for v in m.getVars():
    print(f"{v.varName}: {v.x}")
print(f"Objective Value: {m.objVal}")
```
